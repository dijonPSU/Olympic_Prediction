```python
# Libaries
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
# Read data in
teams = pd.read_csv("teams.csv")

# Get rid of extra columns
teams = teams[["team", "country", "year", "athletes", "age", "prev_medals", "medals"]]
```

```python
# Data Cleaning
teams = teams.dropna()
```

```python
# Split Data 
trainData = teams[teams["year"] < 2012].copy()
testData = teams[teams["year"] >= 2012].copy()
```

```python
# Set up model
model = LinearRegression()
columns = ["athletes", "prev_medals"]
predicting = "medals"

# Train the model
model.fit(trainData[columns], trainData[predicting])

# Making predictions on test data
testData["predictions"] = model.predict(testData[columns])

# No negative values
testData["predictions"] = testData["predictions"].apply(lambda x: max(x, 0))

# Round predictions
testData["predictions"] = testData["predictions"].apply(round)
```
